```python
# Importing Necessary Libraries for Modeling

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
# Display First Five Rows of the Dataset

df.head()
```

```python
# Preparing Data for Modeling by Dropping Unnecessary Columns

df_model = df.drop(
    ["nameOrig", "nameDest", "isFlaggedFraud"],
    axis=1
)

df_model.head()
```

```python
# Defining Categorical and Numeric Features

categorical = ["type"]

numeric = [
    "amount",
    "oldbalanceOrg",
    "newbalanceOrig",
    "oldbalanceDest",
    "newbalanceDest"
]
```

```python
# Splitting Features and Target Variable

y = df_model["isFraud"]
X = df_model.drop("isFraud", axis=1)
```

```python
# Splitting Data into Training and Testing Sets with Stratification

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
# Creating Preprocessing Pipeline

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric),
        ("cat", OneHotEncoder(drop="first"), categorical)
    ],
    remainder="drop"
)
```

```python
# Creating Modeling Pipeline

pipeline = Pipeline(
    steps=[
        ("prep", preprocessor),
        ("clf", LogisticRegression(
            class_weight="balanced",
            max_iter=1000
        ))
    ]
)
```

```python
# Fitting the Pipeline to the Training Data

pipeline.fit(X_train, y_train)
```

```python
# Making Predictions on the Test Set

y_pred = pipeline.predict(X_test)
```

```python
# Evaluating Model Performance

print(classification_report(y_test, y_pred))
```

```python
# Displaying the Confusion Matrix

confusion_matrix(y_test, y_pred)
```

```python
# Calculating and Displaying the Accuracy Score

pipeline.score(X_test, y_test) * 100
```
